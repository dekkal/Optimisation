```python
import numpy as np
from numpy import linalg as la
import matplotlib.pyplot as plt
```

```python
eps = 0.001
itmax = 10000
N = 100
a = 0. 
B = 1.
alpha = 1.
beta = 2.718281828
h = 1. / (N+1)
hh = h*h                                                  
x0 = np.zeros(N)

u0=np.zeros(N)
```

```python
ap=(-2-hh)/hh
aw=1/hh
c=1/hh
```

```python
def jacobi(N, itmax):
    
    b=np.zeros(N)
    res=np.zeros(N)
    b[0]=-1/hh
    b[N-1]=-2.718281828/hh
    x=np.zeros(N)                
    it=0
    err=1

    while err > eps and it < itmax:
        for i in range (N):
               if i ==0:
                 res[i] = -1/hh-(ap*x[i]+1/hh*x[i+1])
               if (i>0 and i<N-1 ):
                 res[i]=-  (aw*x[i-1]+ap*x[i]+aw*x[i+1])
               if(i==N-1):
                 res[i]=((-beta)/hh)-(aw*x[N-2]+ap*x[N-1])  
        for i in range (N):
            x[i]= x[i]+res[i]/ap
        #print(it)
        err=np.max(abs(res))
        it=it+1
        print("nombre d iteration est:",it) 
    return x
```

```python
x=jacobi( N, itmax)
print(x)

print("la solution approchee est",x)
```

```python
plt.plot(x)
plt.legend()

plt.show()
```
